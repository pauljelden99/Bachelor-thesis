```python
import numpy as np
```

```python
#generate transformation matrix to the BK basis with n orbitals
def gen_beta(n):
    #generate beta matrix of size 2^k
    res = np.array([[1]])
    k=1
    while True:
        res = np.kron([[1, 0],[0,1]], res)
        for i in range(2**k):
            res[0][i] = 1
        if 2**k >= n:
            break
        else:
            k += 1
    # delete redundant rows
    j = 2**k - n
    to_be_deleted = np.arange(j)
    res = np.delete(res, to_be_deleted, 0)
    res = np.delete(res, to_be_deleted, 1)
    return res

def pi_mat(n):
    res = np.zeros((n, n))
    for i  in range(n):
        for j in range(n):
            if j >= i:
                res[i][j] = 1
    return res

#calculate update set for the i-th qubit with n orbitals 
def update(n, i):
    beta = gen_beta(n)
    k = n-i-1
    res = np.array([])
    
    # get index of all rows, that are store a 1 at index i
    for j in range(k):
        if beta[j][k] == 1:
            res = np.append(res, np.array([n-j-1]))
    return res

#calculate determination set for the i-th qubit with n orbitals 
def determination(n, i):
    beta = gen_beta(n)
    inverse = np.linalg.inv(beta)
    res = np.array([])
    # get index of all columns, that store a 1 at index i 
    for j in range(n):
        if inverse[n-i-1][j] != 0 and j != n-i-1:
            res = np.append(res, np.array([n-j-1]))
    return res

#calculate parity set
def parity(n, i):
    # calculate transformation from BK to parity basis
    beta = gen_beta(n)
    pi = pi_mat(n)
    inverse = np.linalg.inv(beta)
    mat = np.matmul(pi, inverse)
    res = np.array([])
    for k in range(n):
        for j in range(k+1, n):
            if mat[n-j-1][n-k-1] == 0:
                mat[n-j-1][n-k-1] = 1
                break
    # get index of all columns, that store a 1 at index i 
    for j in range(n):
        if mat[n-i-1][j] != 0 and j != n-i-1:
            res = np.append(res, np.array([n-j-1]))
    return res
```
